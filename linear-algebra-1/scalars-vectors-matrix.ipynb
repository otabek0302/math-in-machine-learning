Norms - A norm = length (magnitude) of a vector
It answers: “How big is this vector?”
• L2 → sqrt of squares
• L1 → sum of absolute values
• L∞ → max absolute value

Base Python (manual)

```python
v = [3, 4]

l2 = math.sqrt(sum(x**2 for x in v))
l1 = sum(abs(x) for x in v)
linf = max(abs(x) for x in v)

print(l2)   # 5.0
print(l1)   # 7
print(linf) # 4
```

NumPy

```python
v = np.array([3, 4])

print(np.linalg.norm(v))        # L2
print(np.linalg.norm(v, ord=1)) # L1
print(np.linalg.norm(v, ord=np.inf)) # L∞
```

PyTorch

```python
v = torch.tensor([3.0, 4.0])

print(torch.norm(v, p=2))      # L2
print(torch.norm(v, p=1))      # L1
print(torch.norm(v, p=float('inf')))  # L∞
```

TensorFlow

```python
v = tf.constant([3.0, 4.0])

print(tf.norm(v, ord=2))        # L2
print(tf.norm(v, ord=1))        # L1
print(tf.norm(v, ord=float('inf')))  # L∞
```

Squared L2 Norm A Squared L2 Norm = sum of squares of vector elements (no square root)

```python
l2 = [3, 4]
l2_np = np.array([3, 4], dtype=np.float32)
l2_th = th.tensor([3, 4], dtype=th.float32)
l2_tf = tf.constant([3, 4], dtype=tf.float32)

# Base Python
squared_l2 = sum(x**2 for x in l2)
print(squared_l2)  # 25

# NumPy
print(np.linalg.norm(l2_np)**2)
print(np.sum(l2_np**2))
print(np.dot(l2_np, l2_np))

# PyTorch
print(th.norm(l2_th)**2)
print(th.sum(l2_th**2))
print(th.dot(l2_th, l2_th))

# TensorFlow
print(tf.norm(l2_tf)**2)
print(tf.reduce_sum(l2_tf**2))
print(tf.tensordot(l2_tf, l2_tf, axes=1))
```

Basis, Orthogonal, and Orthonormal Vectors
Basis - A basis = a set of vectors that can build (represent) any vector in a space.
Standard basis: e1 = [1, 0] | e2 = [0, 1]
Any vector: v = [3, 4] = 3·e1 + 4·e2

Orthogonal Vectors - Two vectors are orthogonal if they are perpendicular
Condition: v⋅w = 0Example: v = [1, 0]  | w = [0, 1]
Dot product:  1*0 + 0*1 = 0 → orthogonal

Orthonormal Vectors -  orthogonal (perpendicular) AND each vector has norm = 1Example: v = [1, 0]  | w = [0, 1]
dot = 0 → orthogonal → norm = 1 → unit  → orthonormal

Tiny ML connection
Orthonormal vectors → used in embeddings, PCA
Basis → feature representation
Orthogonality → reduces redundancy

```python
# NumPy
v = np.array([1, 0])
w = np.array([0, 1])

# dot product
dot = np.dot(v, w)

# norms
norm_v = np.linalg.norm(v)
norm_w = np.linalg.norm(w)

print(dot)     # 0 → orthogonal
print(norm_v)  # 1
print(norm_w)  # 1
```

```python
# PyTorch
v = th.tensor([1.0, 0.0])
w = th.tensor([0.0, 1.0])

dot = th.dot(v, w)
norm_v = th.norm(v)
norm_w = th.norm(w)

print(dot)
print(norm_v)
print(norm_w)
```

```python
# TensorFlow
v = tf.constant([1.0, 0.0])
w = tf.constant([0.0, 1.0])

dot = tf.tensordot(v, w, axes=1)
norm_v = tf.norm(v)
norm_w = tf.norm(w)

print(dot)
print(norm_v)
print(norm_w)
```

Vector Operations (addition, scalar multiplication, dot product).

Vector Addition - Add vectors element-wise
v = [1, 2]
w = [3, 4]
v + w = [4, 6]

Scalar Multiplication - Multiply each element by a number
v = [1, 2]
k = 3
k * v = [3, 6]

Dot Product (very important) - Multiply elements → then sum

Formula: V ⋅ W = X1 Y1 +X2 Y2 + … + Xn Yn
v = [1, 2]
w = [3, 4]

1 * 3 + 2 * 4 = 3 + 8 = 11

Meaning
* measures similarity
* if result = 0 → orthogonal
* larger → more aligned

```python
# Base Python
v = [1, 2]
w = [3, 4]

dot = sum(v[i] * w[i] for i in range(len(v)))
print(dot)

# NumPy
v = np.array([1, 2])
w = np.array([3, 4])

dot = np.dot(v, w)
print(dot)

# PyTorch
v = th.tensor([1, 2])
w = th.tensor([3, 4])

dot = th.dot(v, w)
print(dot)

# TensorFlow
v = tf.constant([1, 2])
w = tf.constant([3, 4])

dot = tf.tensordot(v, w, axes=1)
print(dot)
```
