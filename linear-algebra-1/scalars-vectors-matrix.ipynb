Import Libraries

```python
import numpy as np
import torch as th
import tensorflow as tf
import math
```

Scalars (Rank 0 Tensors) in Base Python

```python
x = 5 # if we need more specificity (e.g., int16, uint8), we use NumPy or another numeric lib
y = 2.5
z = -10
```

```python
print(x)
print(type(x))

print(y)
print(type(y))

print(z)
print(type(z))
```

Scalars in PyTorch

- PyTorch and TensorFlow are the two most popular automatic differentiation libraries in Python, itself the most popular programming language in ML.

- PyTorch tensors are designed to be pythonic, i.e., to feel and behave like NumPy arrays.

- The advantage of PyTorch tensors relative to NumPy arrays is that they easily be used for operations on GPU (see here for example).

```python
x = th.tensor(5) # type specification optional, e.g.: dtype=th.float16
y = th.tensor(2.5)
```

```python
print(x)
print(x.ndim)   # number of dimensions
print(x.shape)  # shape of scalar
print("----------")
print(y)
print(y.ndim)
print(y.shape)
```

Scalars in TensorFlow

- As with TF tensors, in PyTorch we can similarly perform operations, and we can easily convert to and from NumPy arrays.

```python
x = tf.constant(5)
y = tf.constant(2.5)
```

```python
print(x)
print(tf.rank(x))
print(x.shape)
print("-----------")
print(y)
print(tf.rank(y))
print(y.shape)
```

Tiny practical example

```python
# Base Python
loss = 0.128
print(loss)
```

```python
# PyTorch
loss = th.tensor(0.128)
print(loss)
print(loss.ndim)
```

```python
#TensorFlow
loss = tf.constant(0.128)
print(loss)
print(tf.rank(loss))
```

Vectors (Rank 1 Tensors) - A vector is an ordered list of numbers.

```python
v = [1, 2, 3]

# Base Python
print(v)
print(type(v))
print(len(v))  # number of elements
```

NumPy vector

```python
v = np.array([1, 2, 3])

# NumPy vector
print(v)
print(v.shape)   # dimensions
print(v.ndim)    # rank
```

PyTorch vector

```python
v = th.tensor([1, 2, 3])

# PyTorch vector
print(v)
print(v.shape)
print(v.ndim)
```

TensorFlow vector

```python
v = tf.constant([1, 2, 3])

# TensorFlow vector
print(v)
print(tf.rank(v))
print(v.shape)
```

Tiny practical example (ML)
Features of one house: [size, rooms, age] = [120, 3, 10]

```python
# NumPy vector
np_house = np.array([120, 3, 10])
print(np_house)

# PyTorch vector
th_house = th.tensor([120, 3, 10])
print(th_house)

# TensorFlow vector
tf_house = tf.constant([120, 3, 10])
print(tf_house)
```
