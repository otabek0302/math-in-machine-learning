```python
# 1. Scalar Operations in ML - A scalar in ML is one number.

# Base Python
# scalar = one number
b = 2.0                 # bias (scalar)
x = 3.0                 # one feature value (scalar)
w = 0.5                 # weight (scalar)

z = w * x + b           # linear model output (scalar)
print(z)                # 3.5

lr = 0.1                # learning rate (scalar)
theta = 1.0             # parameter (scalar)
grad = 0.6              # dL/dtheta (scalar)

theta = theta - lr * grad
print(theta)            # 0.94


# Numpy
import numpy as np

y = np.array([3.0, 5.0, 2.0])
yhat = np.array([2.5, 4.0, 2.2])

mse = np.mean((y - yhat) ** 2)   # scalar loss
print(mse)

p = 0.2
keep_prob = 1.0 - p              # scalar hyperparameter
print(keep_prob)


# PyTorch
import torch

y = torch.tensor([3.0, 5.0, 2.0])
yhat = torch.tensor([2.5, 4.0, 2.2])

loss = torch.mean((y - yhat) ** 2)  # scalar tensor (0-dim)
print(loss.item())                  # Python scalar

x = torch.tensor(2.0, requires_grad=True)  # scalar parameter
loss2 = (x - 5) ** 2                        # scalar loss
loss2.backward()
print(x.grad.item())                        # scalar gradient


# TensorFlow
import tensorflow as tf

y = tf.constant([3.0, 5.0, 2.0])
yhat = tf.constant([2.5, 4.0, 2.2])

mse = tf.reduce_mean(tf.square(y - yhat))  # scalar tensor
print(float(mse.numpy()))

x = tf.Variable(2.0)  # scalar parameter
with tf.GradientTape() as tape:
    loss = (x - 5.0) ** 2                  # scalar loss
grad = tape.gradient(loss, x)
print(float(grad.numpy()))                 # scalar gradient
```

```python
# 2. Vector Operations in ML - A vector is an ordered list of numbers (1D).

# In ML, vectors commonly represent:
# - feature vectors (one data point)
# - weight vectors
# - embeddings
# - gradients

# Base Python (lists)
a = [1, 2, 3]
b = [10, 20, 30]

# element-wise addition
add = [ai + bi for ai, bi in zip(a, b)]
print("a + b =", add)  # [11, 22, 33]

# element-wise multiplication (mask/gate)
mask = [1, 0, 1]
masked = [ai * mi for ai, mi in zip(a, mask)]
print("a * mask =", masked)  # [1, 0, 3]

# dot product (linear score)
w = [0.2, -0.5, 1.0]
x = [3, 4, 2]
dot = sum(wi * xi for wi, xi in zip(w, x))
print("w·x =", dot)  # 0.6


# NumPy (preferred for vector math)
import numpy as np

va = np.array([1.0, 2.0, 3.0])
vb = np.array([10.0, 20.0, 30.0])

print("va + vb =", va + vb)          # element-wise addition
print("va * vb =", va * vb)          # element-wise multiplication
print("va·vb =", va @ vb)            # dot product

# L2 norm (magnitude)
print("||va|| =", np.linalg.norm(va))

# cosine similarity
cos_sim = (va @ vb) / (np.linalg.norm(va) * np.linalg.norm(vb))
print("cos(va, vb) =", cos_sim)

# broadcasting: scale every element by a scalar
print("2 * va =", 2 * va)


# PyTorch (vectors as 1D tensors)
import torch

ta = torch.tensor([1.0, 2.0, 3.0])
tb = torch.tensor([10.0, 20.0, 30.0])

print("ta + tb =", ta + tb)
print("ta * tb =", ta * tb)
print("ta·tb =", torch.dot(ta, tb))
print("||ta|| =", torch.linalg.norm(ta))


# TensorFlow (vectors as 1D tensors)
import tensorflow as tf

fa = tf.constant([1.0, 2.0, 3.0])
fb = tf.constant([10.0, 20.0, 30.0])

print("fa + fb =", (fa + fb).numpy())
print("fa * fb =", (fa * fb).numpy())
print("fa·fb =", tf.tensordot(fa, fb, axes=1).numpy())
print("||fa|| =", tf.linalg.norm(fa).numpy())



```

```python
# 3. Matrix Operations in ML - A matrix is a 2D grid of numbers (rows x columns).

# In ML, matrices commonly represent:
# - weights in a linear layer (W)
# - a batch of data (X): each row is one example
# - covariance/correlation matrices
# - attention score matrices (QK^T)

# Make this cell runnable even if the global import cell wasn't executed.
import numpy as np
import torch
import tensorflow as tf

# Base Python (lists of lists)
A = [[1, 2],
     [3, 4]]
B = [[10, 20],
     [30, 40]]

# matrix addition (element-wise)
A_plus_B = [[A[i][j] + B[i][j] for j in range(2)] for i in range(2)]
print("A + B =", A_plus_B)  # [[11, 22], [33, 44]]

# scalar * matrix
alpha = 0.5
alphaA = [[alpha * A[i][j] for j in range(2)] for i in range(2)]
print("alpha * A =", alphaA)  # [[0.5, 1.0], [1.5, 2.0]]

# matrix transpose
AT = [[A[j][i] for j in range(2)] for i in range(2)]
print("A^T =", AT)  # [[1, 3], [2, 4]]

# matrix multiplication (2x2)
C = [
    [A[0][0]*B[0][0] + A[0][1]*B[1][0], A[0][0]*B[0][1] + A[0][1]*B[1][1]],
    [A[1][0]*B[0][0] + A[1][1]*B[1][0], A[1][0]*B[0][1] + A[1][1]*B[1][1]],
]
print("A @ B =", C)


# NumPy (preferred for matrices)
A = np.array([[1., 2.],
              [3., 4.]])
B = np.array([[10., 20.],
              [30., 40.]])

print("A + B =\n", A + B)
print("alpha * A =\n", 0.5 * A)
print("A @ B =\n", A @ B)      # matrix multiplication
print("A.T =\n", A.T)          # transpose

# element-wise ops (not matrix multiply)
print("A * B (element-wise) =\n", A * B)
print("exp(A) (element-wise) =\n", np.exp(A))

# trace and determinant
print("tr(A) =", np.trace(A))
print("det(A) =", np.linalg.det(A))

# inverse / pseudo-inverse
print("A^{-1} =\n", np.linalg.inv(A))
print("pinv(A) =\n", np.linalg.pinv(A))

# SVD: A = U S V^T (used in PCA / low-rank approximation)
U, S, Vt = np.linalg.svd(A)
print("S (singular values) =", S)

# simple low-rank approximation with k=1
k = 1
A_rank1 = (U[:, :k] * S[:k]) @ Vt[:k, :]
print("rank-1 approx(A) =\n", A_rank1)


# PyTorch
TA = torch.tensor([[1., 2.],
                   [3., 4.]])
TB = torch.tensor([[10., 20.],
                   [30., 40.]])

print("TA + TB =\n", TA + TB)
print("0.5 * TA =\n", 0.5 * TA)
print("TA @ TB =\n", TA @ TB)
print("TA.T =\n", TA.T)
print("TA * TB (element-wise) =\n", TA * TB)
print("trace(TA) =", torch.trace(TA))
print("det(TA) =", torch.det(TA))
print("inv(TA) =\n", torch.linalg.inv(TA))
print("pinv(TA) =\n", torch.linalg.pinv(TA))


# TensorFlow
FA = tf.constant([[1., 2.],
                  [3., 4.]])
FB = tf.constant([[10., 20.],
                  [30., 40.]])

print("FA + FB =\n", (FA + FB).numpy())
print("0.5 * FA =\n", (0.5 * FA).numpy())
print("FA @ FB =\n", tf.matmul(FA, FB).numpy())
print("transpose(FA) =\n", tf.transpose(FA).numpy())
print("FA * FB (element-wise) =\n", (FA * FB).numpy())

# trace / determinant
print("trace(FA) =", tf.linalg.trace(FA).numpy())
print("det(FA) =", tf.linalg.det(FA).numpy())

# inverse / pseudo-inverse
print("inv(FA) =\n", tf.linalg.inv(FA).numpy())
print("pinv(FA) =\n", tf.linalg.pinv(FA).numpy())

# SVD
s, u, v = tf.linalg.svd(FA)  # FA = u @ diag(s) @ v^T
print("singular values =", s.numpy())
```

```python
# 4. Tensor Operations in ML - A tensor is an N-dimensional array of numbers.

# Scalars are 0D, vectors are 1D, matrices are 2D, and tensors generalize to 3D+.
# Common ML shapes:
# - images: (batch, height, width, channels) or (batch, channels, height, width)
# - sequences: (batch, seq_len, dim)
# - attention: (batch, heads, seq_len, seq_len)

import numpy as np
import torch
import tensorflow as tf


# -----------------------------
# PyTorch tensor operations
# -----------------------------
# Create a 3D tensor: (batch=2, seq=3, dim=4)
X = torch.arange(2 * 3 * 4).float().reshape(2, 3, 4)
print("X shape:", X.shape)

# 1) Reshape / View
X_flat = X.view(2, -1)  # flatten last dimensions
print("X_flat shape:", X_flat.shape)

# 2) Transpose / Permute (reorder dimensions)
X_perm = X.permute(0, 2, 1)  # (batch, dim, seq)
print("X_perm shape:", X_perm.shape)

# 3) Squeeze / Unsqueeze (add/remove size-1 dims)
v = torch.tensor([1., 2., 3.])
v_unsq = v.unsqueeze(0)  # (1, 3)
print("v_unsq shape:", v_unsq.shape)
print("squeezed back:", v_unsq.squeeze(0).shape)

# 4) Concat / Stack
A = torch.tensor([[1., 2.], [3., 4.]])
B = torch.tensor([[10., 20.], [30., 40.]])
cat0 = torch.cat([A, B], dim=0)     # (4, 2)
stack0 = torch.stack([A, B], dim=0) # (2, 2, 2)
print("cat dim=0 shape:", cat0.shape)
print("stack dim=0 shape:", stack0.shape)

# 5) Index / Gather (select using indices)
# Example: pick one element from each row
M = torch.tensor([[5., 6., 7.],
                  [8., 9., 10.]])
idx = torch.tensor([[2],
                    [0]])  # per-row index (shape (2,1))
gathered = M.gather(1, idx)
print("gathered:\n", gathered)

# 6) Masking
scores = torch.tensor([[0.1, 0.5, -0.2],
                       [0.0, -1.0, 2.0]])
mask = torch.tensor([[1, 1, 0],
                     [1, 0, 1]]).bool()
masked_scores = scores.masked_fill(~mask, float("-inf"))
print("masked_scores:\n", masked_scores)

# 7) Reduction (sum/mean over dims)
print("mean over seq (dim=1) shape:", X.mean(dim=1).shape)  # (batch, dim)

# 8) Batched Matmul (batch matrix multiply)
# A: (batch, n, m), B: (batch, m, p) -> (batch, n, p)
A_b = torch.randn(2, 3, 4)
B_b = torch.randn(2, 4, 5)
C_b = torch.bmm(A_b, B_b)
print("bmm output shape:", C_b.shape)

# 9) Einstein Summation (flexible contractions)
# Equivalent to batched matmul: (b,i,k) x (b,k,j) -> (b,i,j)
C_e = torch.einsum("bik,bkj->bij", A_b, B_b)
print("einsum output shape:", C_e.shape)


# -----------------------------
# TensorFlow tensor operations
# -----------------------------
X = tf.reshape(tf.range(2 * 3 * 4, dtype=tf.float32), (2, 3, 4))
print("TF X shape:", X.shape)

# reshape
X_flat = tf.reshape(X, (2, -1))
print("TF X_flat shape:", X_flat.shape)

# transpose / permute
X_perm = tf.transpose(X, perm=(0, 2, 1))
print("TF X_perm shape:", X_perm.shape)

# squeeze / expand dims
v = tf.constant([1., 2., 3.])
v_unsq = tf.expand_dims(v, axis=0)
print("TF v_unsq shape:", v_unsq.shape)
print("TF squeezed shape:", tf.squeeze(v_unsq, axis=0).shape)

# concat / stack
A = tf.constant([[1., 2.], [3., 4.]])
B = tf.constant([[10., 20.], [30., 40.]])
cat0 = tf.concat([A, B], axis=0)
stack0 = tf.stack([A, B], axis=0)
print("TF concat shape:", cat0.shape)
print("TF stack shape:", stack0.shape)

# gather
M = tf.constant([[5., 6., 7.],
                 [8., 9., 10.]])
idx = tf.constant([[2],
                   [0]])
gathered = tf.gather(M, idx, axis=1, batch_dims=1)
print("TF gathered:\n", gathered.numpy())

# masking
scores = tf.constant([[0.1, 0.5, -0.2],
                      [0.0, -1.0, 2.0]])
mask = tf.constant([[True, True, False],
                    [True, False, True]])
masked_scores = tf.where(mask, scores, tf.constant(-np.inf, dtype=tf.float32))
print("TF masked_scores:\n", masked_scores.numpy())

# reduction
print("TF mean over seq axis=1 shape:", tf.reduce_mean(X, axis=1).shape)

# batched matmul
A_b = tf.random.normal((2, 3, 4))
B_b = tf.random.normal((2, 4, 5))
C_b = tf.matmul(A_b, B_b)
print("TF matmul output shape:", C_b.shape)

# einsum
C_e = tf.einsum("bik,bkj->bij", A_b, B_b)
print("TF einsum output shape:", C_e.shape)
```

```python
# Performance & Best Practices (Practical Tips)

import torch

# 1) Contiguous memory (PyTorch)
# After permute, memory layout changes. Call .contiguous() before view() in many cases.
X = torch.randn(2, 3, 4)
Xp = X.permute(0, 2, 1)           # not contiguous
print("is_contiguous before:", Xp.is_contiguous())

Xp2 = Xp.contiguous().view(2, -1) # safe
print("view after contiguous shape:", Xp2.shape)

# 2) In-place vs out-of-place
# In-place ops can save memory but can break autograd if misused.
y = torch.randn(5, requires_grad=True)
loss = (y * y).sum()
loss.backward()
print("grad ok:", y.grad.shape)

# 3) Dimension order matters (NCHW vs NHWC)
# PyTorch convs typically prefer NCHW: (batch, channels, height, width)
images_nchw = torch.randn(32, 3, 28, 28)
print("NCHW shape:", images_nchw.shape)

# 4) Avoid Python loops over tensor dims
# Example: sum over last dim (vectorized)
A = torch.randn(1024, 512)
col_sum = A.sum(dim=0)
print("col_sum shape:", col_sum.shape)

# 5) Prefer fused/optimized ops when available
# Example: linear layer uses a fused matmul + bias inside torch.nn.functional.linear
import torch.nn.functional as F
W = torch.randn(128, 512)
b = torch.randn(128)
x = torch.randn(32, 512)
out = F.linear(x, W, b)
print("linear output shape:", out.shape)
```
