Task 9 - Condition Number

Task 10 - Numerical Gradient via Finite Differences

Approximate the gradient of a scalar function using central finite differences:
df/dx_i ≈ (f(x + eps*e_i) - f(x - eps*e_i)) / (2*eps)

Verify that the numerical gradient matches the analytical gradient.

```python
import numpy as np

# f(x) = x0^3 + 2*x1^2 - x0*x1
# df/dx0 = 3*x0^2 - x1
# df/dx1 = 4*x1 - x0

def f(x):
    return x[0]**3 + 2*x[1]**2 - x[0]*x[1]

x = np.array([2.0, 3.0])
eps = 1e-5

# numerical gradient via central finite differences
grad_num = np.zeros(2)
for i in range(2):
    e_i = np.zeros(2); e_i[i] = 1.0
    grad_num[i] = None  # (f(x + eps*e_i) - f(x - eps*e_i)) / (2*eps)

# analytical gradient
grad_anal = np.array([3*x[0]**2 - x[1], 4*x[1] - x[0]])

print(f'numerical : {grad_num.round(6)}')
print(f'analytical: {grad_anal}')
print(f'match: {np.allclose(grad_num, grad_anal, atol=1e-4)}')
```
