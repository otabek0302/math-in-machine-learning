Task 11 - Eigenvalues and Eigenvectors

An eigenvector v of matrix A satisfies: A @ v = lambda * v
Lambda is the eigenvalue -- it scales the eigenvector without changing its direction.
For a symmetric matrix, all eigenvalues are real and eigenvectors are orthogonal.

```python
import numpy as np

A = np.array([[4.0, 2.0],
              [2.0, 3.0]])  # symmetric positive definite

# compute eigenvalues and eigenvectors
eigenvalues, eigenvectors = None  # np.linalg.eig(A)

print(f'A:\n{A}')
print(f'eigenvalues : {eigenvalues}')
print(f'eigenvectors (columns):\n{eigenvectors}')

# verify: A @ v = lambda * v for each eigenvector (column)
for i in range(len(eigenvalues)):
    lam = eigenvalues[i]
    v   = eigenvectors[:, i]
    lhs = A @ v
    rhs = lam * v
    print(f'\neigenvector {i}: {v.round(4)}')
    print(f'  A @ v   = {lhs.round(4)}')
    print(f'  lam * v = {rhs.round(4)}')
    print(f'  match   : {np.allclose(lhs, rhs)}')
```
